```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
meteorite_df = pd.read_csv("meteorite_data.csv")
```

```python
meteorite_df.dropna(inplace=True)
```

```python
meteorite_df.head()
```

```python
# Get all the Earth meteorites that fell before the year 2000

meteorites_fell_before_2000 = meteorite_df[meteorite_df["Year at which Earth Meteorite was hit"].apply(lambda x: int(str(x)[:4])) <=2000]
```

```python
len(meteorites_fell_before_2000)
```

```python
len(meteorite_df)
```

```python
# Pie chart for better visualisation
values = [len(meteorites_fell_before_2000), len(meteorite_df) - len(meteorites_fell_before_2000)]  # Two quantitative values
labels = ['Before 2000', 'After 2000']  # Labels for each value

# Plotting
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Title
plt.title('Pie Chart for meterioites and their year of falling')

# Display the chart
plt.show()
```

```python
# Get all the earth meteorites co-ordinates who fell before the year 1970
meteorites_fell_before_1970 = meteorite_df[meteorite_df["Year at which Earth Meteorite was hit"].apply(lambda x: int(str(x)[:4])) <=1970]
```

```python
meteorites_fell_before_1970[["Name of Earth Meteorite", "reclat", "reclong"]]
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Latitude and Longitude data
latitudes = meteorites_fell_before_1970["reclat"].to_list()
longitudes = meteorites_fell_before_1970["reclong"].to_list()
marks = meteorites_fell_before_1970["Name of Earth Meteorite"].to_list()

# Load the world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a new plot
fig, ax = plt.subplots(figsize=(20, 12))

# Plot the world map
world.plot(ax=ax, color='lightgray', edgecolor='black')

# Plot marks at the specified coordinates
ax.scatter(longitudes, latitudes, color='red', marker='o', s=20)

# Add labels to each mark
for i, mark in enumerate(marks):
    ax.text(longitudes[i], latitudes[i], mark, fontsize=4, va='center', ha='center')

# Show the map
plt.title('World Map showing meteor hits before 1970')
plt.show()
```

```python
# Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg
heavy_meteors = meteorite_df[meteorite_df["Mass of Earth Meteorite"] > 10000.0]
```
