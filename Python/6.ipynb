```python
import pandas as pd
import matplotlib.pyplot as plt

pokemon_df = pd.read_csv("pokemon_data.csv")
pokemon_df.head()
```

```python
# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_less_than_5 = pokemon_df[pokemon_df['spawn_chance'] < 0.05]
plt.figure(figsize=(16, 8))
plt.bar(spawn_rate_less_than_5['name'], spawn_rate_less_than_5['spawn_chance'])
plt.xlabel('Pokmon')
plt.ylabel('Spawn Chance')
plt.title('Pokmon with Spawn Rate Less Than 5%')
plt.xticks(rotation=90)
plt.show()

```

```python
# Plot for all Pokemons that have less than 4 weaknesses:
    
less_than_4_weaknesses = pokemon_df[pokemon_df['weaknesses'].apply(lambda x: len(x.split(','))) < 4]

plt.figure(figsize=(20, 8))
plt.bar(less_than_4_weaknesses['name'], less_than_4_weaknesses['weaknesses'].apply(lambda x: len(x)))
plt.xlabel('Pokmon')
plt.ylabel('Number of Weaknesses')
plt.title('Pokmon with Less Than 4 Weaknesses')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot for all Pokemons that have no multipliers at all:

import matplotlib.pyplot as plt

# Count the frequencies of Pokemons with 0 multipliers and more than 0 multipliers
zero_multipliers_freq = len(pokemon_df[pokemon_df['multipliers'] == '0'])
more_than_zero_multipliers_freq = len(pokemon_df[pokemon_df['multipliers'] != '0'])

# Create the bar graph
plt.figure(figsize=(6, 4))
plt.bar(['0 Multipliers', 'More than 0 Multipliers'], [zero_multipliers_freq, more_than_zero_multipliers_freq])
plt.xlabel('Multipliers')
plt.ylabel('Frequency')
plt.title('Frequency of Pokemons with 0 Multipliers and More than 0 Multipliers')
plt.show()

```
